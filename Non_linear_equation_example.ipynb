# Effect of initial guess and system parameters on the convergence of iterative methods

The purpose of this notebook is to ilustrate how iterative methods in solving optimization problems are sensetive to intitial condition or guess.

```python
def iterfunc(r,x_n0):            #Define the iterative formuation
  x_n1 = r*x_n0*(1-x_n0)
  return x_n1
```

```python
num_iter = 100                             # Total number of iterations
x_list = []                                # Create an empty list to store the results
n_list = [n for n in range(num_iter)]      # Create a list of iteration index (number)
x0 = 0.30001                               # Initial guess for x
r = 2.8                                    # Equation parameter
x = x0

for n in n_list:                           # Main loop for calculating the next x
  x_list.append(x)                         # Add the calculated x to the list
  x_next = iterfunc(r , x)                 # Calculate the next x
  x = x_next                               # Set x for the next iteration
  
```

```python
print(x_list)                  # Print x iterations
```

# Plot the result

```python
import matplotlib.pyplot as plt                           #Import the plotting library
import seaborn as sns                                     #Import Seaborn library more beutiful visualization!!!
sns.set_style("darkgrid")                                 #Set the visualization style, you can also check the "whitegrid" and see the difference
fig = plt.figure(figsize=(12,6))                          #Set the size of the figure
plt.plot(n_list, x_list)            
plt.ylabel('$x_n$', fontsize=12)                          #Set the label for vertical axis
plt.xlabel('$n$ (Iteration counter)', fontsize=12)        #Set the label for horizental axis
plt.show()                                                #Plot!!!
```

# Compare different cases

```python
num_iter = 100
n_list = [n for n in range(num_iter)]

Param_dict = {"Case1": [2.8 , 0.3] , "Case2": [3.3 , 0.30001]}                        # Define different cases as case = [r,x0]

fig = plt.figure(figsize=(12,6))

for case in Param_dict.keys():
  x_list = []
  r = Param_dict[case][0]
  x0 = Param_dict[case][1]
  x = x0
  for n in n_list:
    x_list.append(x)
    x_next = iterfunc(r , x)
    x = x_next

  plt.plot(n_list, x_list, label = f'{case}: $r$={r} , $x_0$={x0}')


plt.legend(fontsize=12)
plt.ylabel('$x_n$', fontsize=12)
plt.xlabel('$n$ (Iteration counter)', fontsize=12)
plt.show()
```
