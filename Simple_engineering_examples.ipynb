Simulate the flow of water through a pipe with a circular cross-section, with a diameter of 0.1 m and a length of 20 m. The pipe is filled with water at a temperature of 20 °C and a pressure of 10^5 Pa. The pressure drop along the length of the pipe is to be determined.

The following code uses the diameter, length, viscosity and density of water to calculate the Reynold's number which is used to determine the type of flow (laminar or turbulent). Then it uses the appropriate equation (Poiseuille's equation for laminar flow and Darcy-Weisbach for turbulent flow) to calculate the pressure drop along the length of the pipe and print the result to the console.

```python
# Import necessary modules
import math

# Define inputs
d = 0.1  # diameter of pipe (m)
L = 20  # length of pipe (m)
v = 0.001  # kinematic viscosity of water (m^2/s)
rho = 997  # density of water (kg/m^3)
g = 9.81  # acceleration due to gravity (m/s^2)

# Calculate Reynold's number
Re = 4 * 0.001 / (math.pi * d * 0.000001)

# Determine type of flow
if Re < 2300:
    print("Laminar flow")
    # Calculate pressure drop using Poiseuille's equation
    deltaP = (8 * rho * L * v) / (math.pi * d**4)
else:
    print("Turbulent flow")
    # Calculate pressure drop using Darcy-Weisbach equation
    f = 0.25 / ((math.log10(3.7 * d/ (0.00001 * L)))**2)
    deltaP = f * (L/d) * (rho * v**2) / 2

# Print result
print("Pressure drop: ", deltaP, " Pa")
```

Python code to solve the problem and plot the result:

In this example, we first import the necessary libraries: matplotlib to plot the results, and numpy to work with arrays. Next, we create an array x which represents the positions along the beam. We then use displacement function to calculate the displacement along the beam length. After this, we plot the displacement along the beam length using the matplotlib library.


```python
import matplotlib.pyplot as plt
import numpy as np

# Define the properties of the beam
E = 200000 # Young's modulus of the beam material (Pa)
I = 0.00001 # Second moment of area of the beam (m^4)
L = 2 # Length of the beam (m)
P = 1000 # Point load at the free end of the beam (N)
x = np.linspace(0, L, 100) # Discretize the beam length for plotting

# Define the displacement function
def displacement(x, L, P, E, I):
    """
    This function calculates the displacement of a cantilever beam under a point load
    :param x: position along the beam (m)
    :param L: length of the beam (m)
    :param P: point load at the free end of the beam (N)
    :param E: Young's modulus of the beam material (Pa)
    :param I: Second moment of area of the beam (m^4)
    :return: displacement at position x (m)
    """
    return (P*x**2)/((6*E*I)*(3*L-x))

# Calculate the displacement along the beam length
y = displacement(x, L, P, E, I)

# Plot the displacement along the beam length
plt.plot(x, y)
plt.xlabel('Position along the beam (m)')
plt.ylabel('Displacement (m)')
plt.title('Displacement of a cantilever beam under a point load')
plt.show()
```

##Problem 13: Present value
A construction company is considering building a new office building. The building will cost $10 million to construct and will generate $1.5 million in net cash flow each year for 20 years. The company's cost of capital is 10%. What is the present value of the building?

Python code to solve the problem:

This code uses a for loop to calculate the present value of the building project. The loop starts from 0 to n-1, at each iteration it calculates the present value of cash flow for year i by dividing it by (1+r)^(i+1) where r is the discount rate. It adds the present value of cash flow for each year to the present value of the project and finally, it subtracts the cost of the project. The final result is the present value of the building.

```python
# Define inputs
cost = 10e6 # project cost in $
cash_flow = 1.5e6 # annual cash flow in $
n = 20 # number of years
r = 0.1 # discount rate

# Calculate present value
pv = 0
for i in range(n):
    pv += cash_flow / (1 + r)**(i+1)
pv -= cost

# Print present value
print("The present value of the building is: ${:.2f}".format(pv))
```

##Problem 14: Present value of a stream of cash flows
Determine the present value of a stream of cash flows over a period of time, given the cash flows, the discount rate and the number of years.

Python code to solve the problem:

In this example, we define a function called present_value that takes three input parameters: the cash flows (cash_flows), the discount rate (discount_rate) and the number of years (years). The function then calculates the present value of a stream of cash flows using the formula of present value. The function returns the present value of cash flows. The main program then calls the function with the given inputs and prints out the result.

```python
def present_value(cash_flows, discount_rate, years):
    """
    This function calculates the present value of a stream of cash flows
    :param cash_flows: a list of cash flows (dollars)
    :param discount_rate: the discount rate as a decimal (e.g. 0.05 for 5%)
    :param years: number of years
    :return: present value of cash flows (dollars)
    """
    present_value = 0
    for i, cash_flow in enumerate(cash_flows):
        present_value += cash_flow / (1 + discount_rate) ** (i+1)
    return present_value

cash_flows = [1000, 1100, 1200, 1300] # cash flows over a period of 4 years (dollars)
discount_rate = 0.03 # discount rate as a decimal (e.g. 0.05 for 5%)
years = 4 # number of years

pv = present_value(cash_flows, discount_rate, years)

print("The present value of the cash flows is {:.2f} dollars.".format(pv))
```

##Problem 19: Monte Carlo simulation
Use Monte Carlo simulation to estimate the probability of failure of a bridge, given a load distribution and material properties.

Python code to solve the problem:

This code uses the Monte Carlo simulation to estimate the probability of failure of a bridge given the load distribution and material properties. The code defines the number of Monte Carlo samples, mean load, standard deviation of the load and yield strength of the material. Then it runs a loop where it generates random load and random material properties. It checks if load is greater than strength, if it is true it increases the failure count. Finally it calculates the probability of failure and prints it.

This is just one example of how you could use Monte Carlo simulation to estimate the probability of failure of a bridge or any other structure. You can use this approach to estimate the probability of failure for different load distributions and material properties.

```python
import random
import numpy as np

# Define inputs
n_samples = 100000  # number of Monte Carlo samples
load_mean = 170  # mean load (kN)
load_stddev = 2  # load standard deviation (kN)

resistance = 200  #  strength of material (KN)

# Initialize failure count
failures = 0

# Run Monte Carlo simulation
for i in range(n_samples):
    # Generate random load
    load = random.gauss(load_mean, load_stddev)
    # Generate random material properties
    strength = random.gauss(resistance, resistance*0.1)

    # Check for failure
    if load > strength:
        failures += 1

# Calculate probability of failure
p_fail = failures / n_samples

# Print result
print("Probability of failure: {:.2f}%".format(p_fail*100))
```

##Problem 20: Optimize traffic signal timing
Determine the optimal traffic signal timing for a four-way intersection. The intersection has four approaches and the traffic flow on each approach is measured in vehicles per hour. The objective is to minimize the total delay experienced by vehicles at the intersection.

Python code to solve the problem:

This code uses the minimize function from scipy library to minimize the total delay experienced by vehicles at the intersection. The minimize function is a function that minimize a given function. The function defined in objective is the function that should be minimized, in this case, it is the total delay experienced by vehicles at the intersection. The constraints are defined in the constraint1 function, which is the total cycle

time being equal to the sum of the green time for each approach.
The scipy library is a python library for optimization and scientific computing, it provides an efficient way to solve the mathematical optimization problems. The function minimize() is used to find the optimal solution, in this case, the optimal green time for each approach. The results are printed out in the last line of the code.

```python
from scipy.optimize import minimize

# Define inputs
flow = [1000, 800, 600, 700]  # traffic flow on each approach (veh/hr)
cycle_time = 120  # total cycle time (s)
green_time = [30, 40, 50, 40]  # initial green time for each approach (s)

# Define the objective function
def objective(x):
    delay = 0
    for i in range(4):
        delay += (flow[i]/3600)*(cycle_time - x[i])
    return delay

# Define the constraints
def constraint1(x):
    return cycle_time - sum(x)

# Set the bounds
b = (0, cycle_time)
bnds = [b, b, b, b]

# Solve the optimization problem
sol = minimize(objective, green_time, bounds=bnds, constraints={"type": "eq", "fun": constraint1})

print("Optimal green time for each approach: ", sol.x)
```

In this example, the objective function is to minimize the weight of the beam, which is represented by the product of the cross-sectional area (x[0]), the material density (x[1]), and the length of the beam (x[2]). The constraint ensures that the buckling factor of safety is greater than or equal to 2, which is represented by the ratio of the critical load (π²EI/(L²)) to the applied load (P). The minimize function from the scipy.optimize library is used to minimize the objective function subject to the constraint. The solution returned by the minimize function, contains the optimal values of the design variables and other information such as the success of the optimization, the number of iterations, etc.

```python
from scipy.optimize import minimize
import numpy as np

P = 100   #Load
def objective(x):
    """ objective function to minimize """
    return x[0]*x[1]*x[2]

def constraint(x):
    """ constraint : buckling factor of safety >= 2 """
    return (np.pi**2*x[0]/(x[1]*x[2]))/P-2

# Initial guess
x0 = [1, 1, 1]

# Bounds
b = (0.1, 10)
bnds = (b, b, b)

# Constraints
con1 = {'type': 'ineq', 'fun': constraint}
cons = ([con1])

# minimize the objective function
sol = minimize(objective, x0, bounds=bnds, constraints=cons)

print(sol)
```

##Problem 22: Design a truss
Design a simple truss by determining the force in each member of a truss, given the loads and geometry of the truss.

Python code to solve the problem:

```python
import numpy as np

# Define inputs
E = 200e9  # modulus of elasticity (Pa)
A = 0.0002  # cross-sectional area (m^2)
P = 100e3  # load (N)

# Define coordinates of joints
coords = np.array([[0,0], [5,0], [5,3], [0,3], [2.5,6]])

# Define connectivity of joints
connectivity = np.array([[0,1], [1,2], [2,3], [3,0], [0,2], [1,4], [2,4], [3,4]])

# Define loads
loads = np.zeros(len(coords)*2)
loads[5] = -P

# Create global stiffness matrix
K_global = np.zeros((len(coords)*2, len(coords)*2))

# Assemble stiffness matrix
for i in range(len(connectivity)):
    node1 = connectivity[i,0]
    node2 = connectivity[i,1]
    x1,y1 = coords[node1]
    x2,y2 = coords[node2]
    L = ((x2-x1)**2 + (y2-y1)**2)**0.5
    c = (x2-x1)/L
    s = (y2-y1)/L
    k_local = np.array([[c*c, c*s, -c*c, -c*s],
                        [c*s, s*s, -c*s, -s*s],
                        [ -c*c, -c*s, c*c, c*s],
                        [ -c*s, -s*s, c*s, s*s]])*(E*A/L)
    for j in range(2):
        for k in range(2):
            K_global[2*node1+j, 2*node1+k] += k_local[j,k]
            K_global[2*node2+j, 2*node2+k] += k_local[j+2,k+2]
            K_global[2*node1+j, 2*node2+k] -= k_local[j,k+2]
            K_global[2*node2+j, 2*node1+k] -= k_local[j+2,k]

# Solve for nodal displacements
displacements = np.linalg.solve(K_global, loads)

# Solve for member forces
forces = np.zeros(len(connectivity))
for i in range(len(connectivity)):
    node1 = connectivity[i,0]
    node2 = connectivity[i,1]
    x1,y1 = coords[node1]
    x2,y2 = coords[node2]
    L = ((x2-x1)**2 + (y2-y1)**2)**0.5
    c = (x2-x1)/L
    s = (y2-y1)/L
    T = np.array([-c, -s, c, s])
    forces[i] = T@np.array([displacements[2*node1], displacements[2*node1+1], displacements[2*node2], displacements[2*node2+1]])

print(forces)
```

## Problem 23: Simulate flow of traffic
Simulate the flow of traffic on a highway using a cellular automata model. The simulation should model the behavior of vehicles on a highway with one lane and a maximum speed of 20 m/s.

The following code use the Intelligent Driver Model (IDM) to simulate the flow of traffic on a highway with one lane and a maximum speed of 20 m/s. The code uses the density of vehicles, maximum speed, and length of the highway to initialize the simulation, then it runs a loop which updates the positions and velocities of the vehicles at each time step. The code uses the Intelligent Driver Model to update the velocities of the vehicles based on the distance to the next vehicle.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define inputs
L = 1000  # length of highway (m)
v_max = 20  # maximum speed (m/s)
density = 0.1  # vehicle density (vehicles/m)
timestep = 10  # simulation timestep (s)

# Initialize simulation
N = int(density * L)  # number of vehicles
positions = np.random.rand(N) * L  # initial vehicle positions (m)
velocities = np.random.rand(N) * v_max  # initial vehicle velocities (m/s)

# Simulation loop
for t in range(10):
    # Update vehicle positions
    positions += velocities * timestep

    # Update vehicle velocities
    for i in range(N):
        # Calculate distance to next vehicle
        distance_ahead = float("inf")
        for j in range(N):
            if j != i:
                distance = positions[j] - positions[i]
                if distance > 0 and distance < distance_ahead:
                    distance_ahead = distance

        # Update velocity according to the Intelligent Driver Model
        v_desired = min(v_max, distance_ahead)
        acceleration = 1  # constant acceleration
        velocities[i] = min(velocities[i] + acceleration * timestep, v_desired)
   
    # Creat highway raster
    highway = np.zeros([L,2])
    for i in range(len(highway)):
      if i in positions.astype(int):
        highway[i,0] = 1.0


    # Plot the highway
    print('Time step: {} seconds'.format(t*timestep))
    plt.figure(figsize=(12, 1))
    plt.imshow(highway.T, cmap='gray_r',  aspect=6)
    plt.axis('off')
    plt.show()
    print('')

```

##Problem 24: Time-temperature response of a concrete slab
Determine the time-temperature response of a concrete slab under fire conditions. The slab has a thickness of 0.2m, a thermal conductivity of 1W/m.K, and a heat transfer coefficient of 10W/m^2.K. The initial temperature of the slab is 20C and the ambient temperature is 800C.

Python code to solve the problem:

This code uses basic mathematical formulas to calculate the time-temperature response of the concrete slab under fire conditions. It uses the numpy library to perform mathematical operations and to create the time array. The code then uses the matplotlib library to plot the time-temperature response. The x-axis represents the time and the y-axis represents the temperature.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define inputs
t = 0.2  # thickness of the slab (m)
k = 1  # thermal conductivity of the slab (W/m.K)
h = 10  # heat transfer coefficient (W/m^2.K)
T_init = 20  # initial temperature of the slab (C)
T_inf = 800  # ambient temperature (C)

# Calculate the time constant
tau = t/(h+k)

# Calculate the time-temperature response
time = np.linspace(0, 1, 100)
T = T_init + (T_inf - T_init)*(1-np.exp(-time/tau))

# Plot the time-temperature response
plt.plot(time, T)
plt.xlabel('Time (s)')
plt.ylabel('Temperature (C)')
plt.title('Time-temperature response')
plt.show()
```

Compare the cost and environmental impact of different pavement materials for a road project. The materials considered are asphalt, concrete and permeable pavement.
Python code to solve the problem:
This code uses a simple decision-making approach by creating a cost score for each material, by adding the environmental impact (multiplied by a factor) to the sum of construction and life-cycle costs. It then sorts the scores and determines the best material. The matplotlib library is used to plot the results.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define inputs
materials = ["Asphalt", "Concrete", "Permeable"]
cost = [500000, 800000, 600000] # Construction cost (USD)
life_cycle_cost = [600000, 900000, 700000] # Life-cycle cost (USD)
environmental_impact = [2, 3, 1] # Environmental impact (1-5 scale)

# Decision making
scores = []
for i in range(len(materials)):
    score = cost[i] + life_cycle_cost[i] + environmental_impact[i]*100_000
    scores.append([materials[i], score])

scores = sorted(scores, key=lambda x: x[1], reverse=False)
best_material = scores[0][0]

# Plotting the results
x = [material for material, _ in scores]
y = [score for _, score in scores]
plt.bar(x, y)
plt.xlabel('Materials')
plt.ylabel('Scores')
plt.title('Decision making for pavement material')
plt.show()

print("The best material for the pavement is: ", best_material)
```
